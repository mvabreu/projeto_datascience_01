```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# importando dataframe total de internações/mês por unidade da Federação
df_internacoes = pd.read_csv('internacoes.csv', encoding="ISO-8859-1", skiprows= 3, sep=";", skipfooter=12, thousands=".", decimal=",")
```

```python
# Excluindo a coluna total do dataframe
df_internacoes = df_internacoes.drop('Total', axis=1)
df_internacoes = df_internacoes.drop('2021/Set', axis=1)
```

```python
df_internacoes['Unidade da Federação'] = df_internacoes['Unidade da Federação'].str.slice(start=3)
```

```python
df_internacoes = df_internacoes.set_index("Unidade da Federação")
```

```python
df_internacoes.head()
```

```python
internacoes_centro_oeste = pd.DataFrame(df_internacoes[23:])
internacoes_centro_oeste
```

```python
# importando dataframe valor total gasto por mês em internações por unidade da Federação
df_valor_total = pd.read_csv('valor_total_internacoes.csv', encoding="ISO-8859-1", skiprows= 3, sep=";", skipfooter=12, thousands=".", decimal=",")
```

```python
# Excluindo a coluna total do dataframe
df_valor_total = df_valor_total.drop('Total', axis=1)
df_valor_total = df_valor_total.drop('2021/Set', axis=1)
```

```python
df_valor_total['Unidade da Federação'] = df_valor_total['Unidade da Federação'].str.slice(start=3)
```

```python
df_valor_total = df_valor_total.set_index("Unidade da Federação")
```

```python
df_valor_total.head()
```

```python
valor_gasto_centro_oeste = pd.DataFrame(df_valor_total[23:])
valor_gasto_centro_oeste
```
