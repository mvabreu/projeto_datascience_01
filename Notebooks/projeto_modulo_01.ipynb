```python
#importando blibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

## Análises/ Tratamento de  Internações no período

```python
# definindo variavel para url do arquivo de dataset internações 
url_1 = 'https://raw.githubusercontent.com/mvabreu/projeto_datascience_01/main/Dados/internacoes.csv'
```

```python
# importando dataframe total de internações/mês por unidade da Federação
df_internacoes = pd.read_csv(url_1, encoding="ISO-8859-1", skiprows= 3, sep=";", skipfooter=12, thousands=".", decimal=",")
```

```python
# Excluindo a coluna total do dataframe
df_internacoes = df_internacoes.drop('Total', axis=1)
df_internacoes = df_internacoes.drop('2021/Set', axis=1)
```

```python
# Tratando coluna Unidade da Federação
df_internacoes['Unidade da Federação'] = df_internacoes['Unidade da Federação'].str.slice(start=3)
```

```python
# Definindo a coluna Unidade da Federação como index do dataframe
df_internacoes = df_internacoes.set_index("Unidade da Federação")
```

```python
df_internacoes.head()
```

```python
# Selecionando estados da Região Centro-Oeste
internacoes_centro_oeste = pd.DataFrame(df_internacoes[23:])
internacoes_centro_oeste
```

## Análises/ Tratamento do Valor Gasto no período


```python
# definindo variavel para url do arquivo de dataser total_gasto
url_2 = 'https://raw.githubusercontent.com/mvabreu/projeto_datascience_01/main/Dados/valor_total_internacoes.csv'
```

```python
# importando dataframe valor total gasto por mês em internações por unidade da Federação
df_valor_total = pd.read_csv(url_2, encoding="ISO-8859-1", skiprows= 3, sep=";", skipfooter=12, thousands=".", decimal=",")
```

```python
# Excluindo a coluna total do dataframe
df_valor_total = df_valor_total.drop('Total', axis=1)
df_valor_total = df_valor_total.drop('2021/Set', axis=1)
```

```python
# Tratando coluna Unidade da Federação
df_valor_total['Unidade da Federação'] = df_valor_total['Unidade da Federação'].str.slice(start=3)
```

```python
# Definindo a coluna Unidade da Federação como index do dataframe
df_valor_total = df_valor_total.set_index("Unidade da Federação")
```

```python
df_valor_total.head()
```

```python
# Selecionando estados da Região Centro-Oeste
valor_gasto_centro_oeste = pd.DataFrame(df_valor_total[23:])
valor_gasto_centro_oeste
```

## Gráficos


```python
ax = internacoes_centro_oeste.head().T.plot(figsize=(10,7), marker='o')

ax.legend(loc='center left', bbox_to_anchor=(1, 0.9))
ax.xaxis.grid()
plt.title('Total de internações por mês - Região Centro-Oeste')
ax.yaxis.grid()
```

```python
ax = valor_gasto_centro_oeste.head().T.plot(figsize=(10,7), marker='o')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.9))
ax.xaxis.grid()
ax.yaxis.grid()
plt.title('Total gasto por mês (Milhões) - Região Centro-Oeste')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
```
